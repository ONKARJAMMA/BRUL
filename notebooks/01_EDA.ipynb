```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/train_FD001.csv')

# Show structure
print(df.head())
print(df.info())

# Plot sensor degradation for 1 unit
unit1 = df[df['unit'] == 1]
plt.plot(unit1['cycle'], unit1['sensor2'], label='Sensor 2')
plt.plot(unit1['cycle'], unit1['sensor3'], label='Sensor 3')
plt.title("Degradation Over Time - Unit 1")
plt.xlabel("Cycle")
plt.ylabel("Sensor Value")
plt.legend()
plt.show()
```
